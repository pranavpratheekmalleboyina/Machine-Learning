```python
#!/usr/bin/env python
# coding: utf-8

# In[58]:


import matplotlib.pyplot as plt
import csv


# In[59]:


X_Data = []
Y_Data = []

#classes = {1:[1,1],2:[1,0],3:[0,1],4:[0,0]}
with open('/preData.csv', 'r') as f: # you have to manually upload dataset before you can run the code
  reader = csv.reader(f)
  cnt = 0
  for row in reader:
    cnt += 1
    print(cnt)
    if cnt == 1:
        continue  # for eleminating the first row from the TestingData list
    else:
        newList = []
        for ele in row:
            newList.append(float(ele))
        print(newList)
        X_Data.append(newList[:-2])
        if newList[-2] == 1 and newList[-1] == 1:
            Y_Data.append(1)
        elif newList[-2] == 1 and newList[-1] == 0:
            Y_Data.append(2)
        elif newList[-2] == 0 and newList[-1] == 1:
            Y_Data.append(3)
        elif newList[-2] == 0 and newList[-1] == 0:
            Y_Data.append(4)

print(len(X_Data[16]),X_Data[16],Y_Data[16])



# In[60]:


from sklearn.model_selection import train_test_split

# Assuming X is your feature data, and y is your labels
X_train, X_temp, Y_train, Y_temp = train_test_split(X_Data, Y_Data, test_size=0.3, random_state=42)
X_val, X_test, Y_val, Y_test = train_test_split(X_temp, Y_temp, test_size=0.5, random_state=42)


# In[61]:


print(len(X_train),len(Y_train),len(X_test))


# In[62]:


import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Assuming you have 14 inputs and 4 outputs
input_size = 16
output_size = 1

# Create a Sequential model
model = Sequential()

# Add a hidden layer with, for example, 64 neurons and ReLU activation function
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))

# Add another hidden layer if needed
# model.add(Dense(32, activation='relu'))

# Add the output layer with 4 neurons (for 4 outputs) and a suitable activation function
model.add(Dense(output_size, activation='softmax'))  # Use 'softmax' for multi-class classification

model.build(input_shape=(None, input_size))

loss=[]

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=["accuracy","AUC","Recall","Precision"])

# Print a summary of the model architecture
model.summary()


# In[63]:


#model.compile(optimizer="adam",loss='categorical_crossentropy',metrics=["accuracy","AUC","Recall","Precision"])

history=model.fit(X_train,Y_train,epochs=30)
plt.plot(history.history['accuracy'],color='green')
plt.plot(history.history['loss'],color='red')
plt.plot(history.history['auc'])
plt.plot(history.history['recall'])
plt.plot(history.history['precision'])
plt.title('model accuracy and loss')

plt.ylabel(' accuracy,loss,auc,recall and precision')
plt.xlabel('epoch')
plt.legend(['accuracy','loss','auc','recall','precision'])
plt.show()
print("test:")
print(model.evaluate(X_test,Y_test))


# In[ ]:
```
