```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=2.0,autopct='%1.1f%%',startangle=75)
plt.axis("equal")
plt.savefig("pranavpiechart.png")
```
