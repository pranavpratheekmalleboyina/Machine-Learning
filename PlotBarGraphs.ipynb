```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))
```

```python
plt.bar(company,revenue)
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
```

```python
profit=[40,2,34,12]
plt.xlabel("Company name")
plt.ylabel("Revenue")
plt.title("US company stock prices")
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,width=0.8,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.2,label="Profit")
plt.legend(loc="best")
```

```python
plt.xlabel("Company name")
plt.ylabel("Revenue")
plt.title("US company stock prices")
ypos = np.arange(len(company))
plt.yticks(ypos,company)
plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend(loc="best")
```
