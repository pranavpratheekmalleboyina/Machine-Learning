```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8)
```

```python
plt.hist(blood_sugar, rwidth=0.8,bins=4)
```

```python
plt.xlabel("Blood Sugar Level")
plt.ylabel("No. of patients")
plt.title("Blood sugar prediction")
plt.hist(blood_sugar, rwidth=0.8,color='red',bins=[100,120,130,140])
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_men,blood_sugar_women],bins=[100,120,130,140],rwidth=0.8,label=['men','women'])
plt.legend()
```
