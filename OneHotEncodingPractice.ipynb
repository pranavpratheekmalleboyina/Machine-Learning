```python
import pandas as pd
```

```python
df = pd.read_csv("homeprices_ohe.csv")
df
```

```python
dummies=pd.get_dummies(df['town']).astype(int)
#df['town'] = df['town'].astype(int)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town','west windsor'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
AREA = final.drop(['price'],axis='columns') #the dataset to contain area
AREA
```

```python
PRICE = final.price
PRICE
```

```python
model.fit(AREA,PRICE)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(AREA,PRICE)
```
